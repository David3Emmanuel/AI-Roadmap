## Differentiation Basics
For most functions, we can compute derivatives using the limit definition:
$$
f'(x) = \lim_{h \to 0} \frac{f(x + h) - f(x)}{h}
$$
For simple functions, we can derive rules like the power rule, product rule, and chain rule to compute derivatives more easily.

```python
def f(x):
    return x**2 + 3*x + 2

def f_prime(x):
    return 2*x + 3

f_prime(1)
```

## Numerical Differentiation
Numerical differentiation approximates the derivative using finite differences. A common method is the central difference:
$$
f'(x) \approx \frac{f(x + h) - f(x - h)}{2h}
$$
where \( h \) is a small number.

```python
def numerical_derivative(f, x, h=1e-5):
    return (f(x + h) - f(x - h)) / (2 * h)

def f(x):
    return x**2 + 3*x + 2

def f_prime_numerical(x):
    return numerical_derivative(f, x)

f_prime_numerical(1)
```

Cons:
- Sensitive to the choice of \( h \)
- Can be inaccurate for very small or very large \( h \)
- Error due to truncation and round-off
- Computationally expensive for high-dimensional inputs

## Symbolic Differentiation
Symbolic differentiation uses algebraic manipulation to find the derivative of a function. Libraries like SymPy can perform symbolic differentiation.


```python
import sympy as sp
from sympy.abc import x

f = x**2 + 3*x + 2
f_prime = sp.diff(f, x)

f_prime
```

```python
f_prime.evalf(subs={x: 1})
```

Cons:
- Expression swell: intermediate expressions can become very large
  $$
  \text{Example: } f(x) = (x^2 + 1)^5 \implies f'(x) = 10x(x^2 + 1)^4
  $$
- Not always efficient for numerical evaluation
- Restricted to closed-form expressions

## Automatic Differentiation (Autodiff)
Autodiff computes derivatives using the chain rule applied to elementary operations.

Pros:
- Same accuracy as symbolic differentiation
- Efficient for numerical evaluation
- Can handle complex functions and control flow

It works in two modes:
1. [Forward Mode](Forward_Mode.ipynb)
2. [Reverse Mode](Reverse_Mode.ipynb)

_The implementations above are extremely simplified and work only for scalar values_
