### Forward Mode
The function is represented as a graph of operations, and we propagate derivatives through this graph.
For each operation, we compute both the value and the derivative with respect to the input.
So, instead of just computing $y = f(x)$, we also compute $\frac{dy}{dx}$.

The catch is that the function needs to be evaluated once for each derivative we want to compute.
For example, for $f(x, y) = x^2 + xy + y^2$, to compute derivatives with respect to both $x$ and $y$, we need to evaluate the function twice.

**Use Case:** Few inputs, many outputs because a single pass can compute the derivative of all outputs for a given input.

```python
import math


class FValue:
    def __init__(self, val, grad=0.) -> None:
        self.val = val
        self.grad = grad
    
    def __add__(self, other) -> 'FValue':
        if not isinstance(other, type(self)): other = FValue(other)
        val = self.val + other.val
        grad = self.grad + other.grad
        return FValue(val, grad)
    
    def __radd__(self, other) -> 'FValue':
        return self + other
    
    def __sub__(self, other) -> 'FValue':
        if not isinstance(other, type(self)): other = FValue(other)
        val = self.val - other.val
        grad = self.grad - other.grad
        return FValue(val, grad)
    
    def __rsub__(self, other) -> 'FValue':
        if not isinstance(other, type(self)): other = FValue(other)
        return other - self
    
    def __mul__(self, other) -> 'FValue':
        if not isinstance(other, type(self)): other = FValue(other)
        val = self.val * other.val
        grad = self.grad * other.val + self.val * other.grad
        return FValue(val, grad)
    
    def __rmul__(self, other) -> 'FValue':
        return self * other
    
    def __truediv__(self, other) -> 'FValue':
        if not isinstance(other, type(self)): other = FValue(other)
        val = self.val / other.val
        grad = (self.grad * other.val - self.val * other.grad) / (other.val ** 2)
        return FValue(val, grad)
    
    def __rtruediv__(self, other) -> 'FValue':
        if not isinstance(other, type(self)): other = FValue(other)
        return other / self
    
    def __pow__(self, other) -> 'FValue':
        if not isinstance(other, type(self)): other = FValue(other)
        val = self.val ** other.val
        grad = (other.val * self.val ** (other.val - 1)) * self.grad + \
               (self.val ** other.val * math.log(self.val)) * other.grad
        return FValue(val, grad)
    
    def __rpow__(self, other) -> 'FValue':
        if not isinstance(other, type(self)): other = FValue(other)
        return other ** self
    
    def __repr__(self) -> str:
        return f'Value({self.val}, {self.grad})'

def Input(val):
    return FValue(val, 1)
```

_My implementation of `__pow__` takes a logarithm which is not defined for negative bases._

```python
def f(x, y): return (2*x+y)/(5*x-y)

z1 = f(Input(1), FValue(1))
print(z1)
z2 = f(FValue(1), Input(1))
print(z2)
```
