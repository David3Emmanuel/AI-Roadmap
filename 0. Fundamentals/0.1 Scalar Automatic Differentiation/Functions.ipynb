```python
from Forward_Mode import FValue, Input
from Reverse_Mode import RValue, Output
```

```python
import math
from typing import Callable

class Function[T: (FValue, RValue)]:
    def __init__(self, _forward_backward: Callable[[T], tuple[float, float]]) -> None:
        self._forward_backward = _forward_backward
    
    def __call__(self, x: T) -> T:
        self.x = x
        f, dx = self._forward_backward(x)
        
        if isinstance(x, FValue):
            return FValue(f, dx * x.grad)
        elif isinstance(x, RValue):
            new_node = RValue(f)
            new_node._prev = [(x, dx)]
            return new_node
        else:
            raise TypeError("Function can only be called with FValue or RValue")


exp = Function(lambda x: (math.exp(x.val), math.exp(x.val)))
sin = Function(lambda x: (math.sin(x.val), math.cos(x.val)))
cos = Function(lambda x: (math.cos(x.val), -math.sin(x.val)))
log = Function(lambda x: (math.log(x.val), 1/x.val))
```

```python
def f(x, y):
    a = x/y
    b = sin(a)
    c = exp(y)
    d = a-c
    e = b+d
    return e*d

print("Forward")
z1 = f(Input(1), FValue(1))
print(z1)
z2 = f(FValue(1), Input(1))
print(z2)

print("Reverse")
x = RValue(1)
y = RValue(1)
z = Output(f(x,y))
print(x,y,z)
```
